```python
import json
import pandas as pd
from datasets import load_dataset
from tqdm import tqdm
import csv
import sys

csv.field_size_limit(sys.maxsize)
```

```python
kilt_nq_dataset = load_dataset("facebook/kilt_tasks", "nq", split="validation")
```

```python
retrieval_dict = {}

with open("nq_test_larger.tsv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f, delimiter="\t")
    for row in reader:
        retrieval_dict[row["question"]] = row

retrieved_items = {}
not_found_count = 0

for item in kilt_nq_dataset:
    question = item["input"]
    if question in retrieval_dict:
        retrieved_items[question] = retrieval_dict[question]
    else:
        not_found_count += 1

for q, details in list(retrieved_items.items())[:5]:
    print(f"Question: {q}\nRetrieved: {details}\n")

print("Not found in retrieval_dict: ",not_found_count)
```

```python
li_final = list(retrieved_items.values())
```

```python
import csv

def write_dicts_to_tsv(data, filename):
    if not data:
        print("No data provided.")
        return
    
    keys = data[0].keys()
    
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=keys, delimiter='\t')
        writer.writeheader()
        writer.writerows(data)

write_dicts_to_tsv(li_final, "nq_test.tsv")
```
